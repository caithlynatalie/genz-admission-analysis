```python
import pandas as pd

my_columns = ['social_media_hours', 'high_school_gpa', 'volunteer_hours', 'essay_score', 'admission_status']
students = pd.read_csv('genz_college_admission_prediction.csv', usecols=my_columns)

students_clean = students.dropna()

students_clean['social_media_hours'] = pd.to_numeric(students_clean['social_media_hours']).copy()
students_clean['high_school_gpa'] = pd.to_numeric(students_clean['high_school_gpa']).copy()
students_clean['volunteer_hours'] = pd.to_numeric(students_clean['volunteer_hours']).copy()
students_clean['essay_score'] = pd.to_numeric(students_clean['essay_score']).copy()
students_clean['admission_status'] = pd.to_numeric(students_clean['admission_status']).copy()

print("Research Correlations:")
correlations = students_clean.corr()['social_media_hours']
print(f"Social Media vs. GPA: {correlations['high_school_gpa']:.2f}")
print(f"Social Media vs. Volunteer Hours: {correlations['volunteer_hours']:.2f}")
print(f"Social Media vs. Essay Score: {correlations['essay_score']:.2f}")
print(f"Social Media vs. Admission Status: {correlations['admission_status']:.2f}")
print("\n")

bins = [0, 3, 6, float('inf')]
labels = ['Low Usage (0-3h)', 'Medium Usage (3-6h)', 'High Usage (6h+)']
students_clean['social_media_group'] = pd.cut(students_clean['social_media_hours'], bins=bins, labels=labels, include_lowest=True)

print("Percentage of Students in Each Category:")
group_percentage = students_clean['social_media_group'].value_counts(normalize=True) * 100
print(group_percentage.map("{:.2f}%".format))
print("\n")

print("Absolute Count of Admitted vs Not Admitted Students:")
admit_counts = pd.crosstab(students_clean['social_media_group'], students_clean['admission_status'])
admit_counts = admit_counts.rename(columns={0: 'Not Admitted (Count)', 1: 'Admitted (Count)'})
print(admit_counts)
print("\n")

print("Academic Trade-off Summary Table:")
summary_table = students_clean.groupby('social_media_group', observed=False).agg(
    average_gpa=('high_school_gpa', 'mean'),
    average_volunteer_time=('volunteer_hours', 'mean'),
    average_essay_score=('essay_score', 'mean'),
    admission_probability=('admission_status', 'mean')
).reset_index()
print(summary_table.to_string(index=False))
print("\n")
```
